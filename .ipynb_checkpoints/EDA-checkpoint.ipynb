## Visualize few audio files

```python
# understanding the wav file form
# use scipy to get frame rate information of the wav file. 

sample_files = glob.glob('sample_audios/*.wav')

from scipy.io import wavfile
flv = (sample_files[0])
rate, n_samples = wavfile.read(flv)
print("Sampling (frame) rate = ", rate)
print("Total samples (frames) = ", n_samples.shape)
print(n_samples)
```

```python
# compute frame rate information of the wav files using scipy
n_samples = []
for file in sample_files:
    rate, n_sample = wavfile.read(file)
    n_samples.append(n_sample)
```

## Analysing Frame lengths for the train and test set samples

```python
# Get n_samples for the whole train set and append to the train_set df

train_set['n_samples'] = train_set.fname.apply(lambda f: len(wavfile.read('data/audio_train/' + f)[1]))
test_set['n_samples'] = test_set.fname.apply(lambda f: len(wavfile.read('data/audio_test/' + f)[1]))
```

```python
test_set = test_set.drop(['usage', 'license'], axis=1)
```

```python
test_set.head()
```

```python
# Analysing the frame length distribution
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14,6))
train_set.n_samples.hist(bins=120, ax=axes[0])
test_set.n_samples.hist(bins=120, ax=axes[1])
plt.suptitle('Sample Distribution in Train and Test Set', ha='center')
```

From the test set samples, we see that around 5 samples have relatively longer length as compared to the others. We will exclude these from our test set. 

## Labels distribution

A look at how the class labels are distributed

```python
np.unique(train_set.label, return_counts=True)
```

```python
category_group = train_set.groupby(['label', 'manually_verified']).count()
category_group = category_group.unstack()
category_group.columns = ['fname_0', 'fname_1', 'n_samples_0', 'n_samples_1']
category_group.head()
```

```python
plot1 = category_group[['n_samples_0']].sort_values('n_samples_0', ascending = 'False').plot(kind = 'barh', figsize = (10,8))
plot1.set_xlabel("Label")
plot1.set_ylabel("Number of Samples");
```

```python
plot2 = category_group[['n_samples_1']].sort_values('n_samples_1', ascending = 'False').plot(kind = 'barh', figsize = (10,8))
plot2.set_xlabel("Label")
plot2.set_ylabel("Number of Samples")
plot2.set_title('No of samples Manually verified')
```
